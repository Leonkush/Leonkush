```python
import pandas as pd
import datetime as dt
import re
import calendar
df=pd.read_csv("train.csv")
```

```python
df
```

```python
h=[]
m=[]
d=[]
dates=[]
for i in range(len(df)):
  date_time = dt.datetime.strptime(df.datetime[i], '%m/%d/%Y %H:%M')
  date_object=date_time.date()
  dates.append(date_object)
  h.append(date_time.hour)
  m.append(calendar.month_name[date_object.month])
  d.append(list(calendar.day_name)[calendar.weekday(date_object.year, date_object.month, date_object.day)])
df["Hour"]=h
df["Month"]=m
df["Weekday"]=d
df["Date"]=dates
```

```python
df
```

```python
b=df.groupby("Month")["Total_booking"].mean()
b.sort_values(ascending=False)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
out = ohe.fit_transform(df["season"].values.reshape(-1,1)).toarray()
df_out = pd.DataFrame(out,columns=["Fall","Spring","Summer","Winter"])
df_out.head()
df.join(df_out)
df=df.drop(["season"],axis=1)
```

```python
ohe_w = OneHotEncoder()
out = ohe_w.fit_transform(df["weather"].values.reshape(-1,1)).toarray()
df_w = pd.DataFrame(out,columns=["Clear + Few clouds","Heavy Rain + Thunderstorm","Light Snow, Light Rain","Mist + Cloudy"])
df_w.head()
df=df.join(df_w)
df.tail(12)
df=df.drop(["weather"],axis=1)
```

```python
ohe_w = OneHotEncoder()
out = ohe_w.fit_transform(df["Month"].values.reshape(-1,1)).toarray()
df_w = pd.DataFrame(out)
df_w.head()
df=df.join(df_w)
df.tail(12)
df=df.drop(["Month"],axis=1)
```
